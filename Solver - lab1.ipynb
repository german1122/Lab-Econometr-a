d) Sean $y = edociv$, $mascot = x$, $ingreso = z$, $\beta_0^* = \alpha_0$ y $\beta_1^* = \alpha_1$, donde el "verdadero modelo" estaría dado por:

$$y = \alpha_0 + \alpha_1 x + \alpha_2 z + v,$$

Dado lo anterior, calculemos el modelo para el que tenemos datos y sustituyamos el verdadero modelo:

$$\hat{\beta_1} = \frac{\sum (x_i-\bar{x})y_i}{\sum (x_i-\bar{x})^2} = \frac{\sum (x_i-\bar{x})(\alpha_0 + \alpha_1 z_i + \varepsilon_i)}{\sum (x_i-\bar{x})^2} =\alpha_0\frac{\sum (x_i-\bar{x}) }{\sum (x_i-\bar{x})^2} + \alpha_1\frac{\sum (x_i-\bar{x})x_i }{\sum (x_i-\bar{x})^2} + \alpha_2\frac{\sum (x_i-\bar{x})z_i }{\sum (x_i-\bar{x})^2} + \frac{\sum (x_i-\bar{x})v_i }{\sum (x_i-\bar{x})^2}, $$


donde el primer término es evidentemente cero; para el segundo término, se puede mostrar lo siguiente: $\sum (x_i -\bar x)x_i = \sum (x_i -\bar x)^2$; el tercer término es la covarianza muestral entre $x$ y la variable omitida, $z$; el cuarto término es la covarianza muestral entre $x$ y el término de error del "verdadero modelo".

Así, 

$$\hat{\beta_1} = \alpha_1 + \alpha_2\frac{\sum (x_i-\bar{x})z_i }{\sum (x_i-\bar{x})^2} + \frac{\sum (x_i-\bar{x})v_i }{\sum (x_i-\bar{x})^2}$$

Por consiguiente,

$$\mathbb{E}\Big(\hat{\beta_1} \Big| x, z\Big) = \alpha_1 + \alpha_2\frac{\sum (x_i-\bar{x})z_i }{\sum (x_i-\bar{x})^2} + \frac{\sum (x_i-\bar{x})\mathbb{E}(v_i) }{\sum (x_i-\bar{x})^2} = \alpha_1+\alpha_2\frac{\sum (x_i-\bar{x})z_i }{\sum (x_i-\bar{x})^2}$$

Así, ya que el coeficiente de $x$ es $\alpha_1$, el sesgo de el estimador $\beta_1$ es el siguiente:

$$S(\hat{\beta_1}) = \mathbb{E}\Big(\hat{\beta_1} \Big| x, z\Big) - \alpha_1 = \alpha_2\frac{\sum (x_i-\bar{x})z_i }{\sum (x_i-\bar{x})^2} = \alpha_2 \frac{\widehat{Cov(x, z)}}{\widehat{Var(x)}}$$

Así, el estimador de $\beta_1$ será insesgado sólo si la covarianza entre $x$ y $z$ es cero, claro ejemplo de incumplimiento del supuesto de exogeneidad. Es decir, en el primer modelo, $x$ está correlacionado con el error mediante $z$, ya que a $z$ lo sacamos del error.
